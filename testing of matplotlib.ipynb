1. Line plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('yahoo.csv')
df.head()
```

```python
# how to plot graph
plt.plot(df['Close'])
```

```python
# if we have to do changes in our graph
plt.plot(df['Close'],color = 'red',marker = ".", 
linewidth = 1 , markersize = 0.2, )
plt.xlabel('days')
plt.ylabel('prise')
plt.title('nifty50 chart')
plt.show()
```

```python
# we can plot graph in graph
plt.plot(df['Close'],color = 'red',marker = ".", 
linewidth = 1 , markersize = 0.2,label = ('close values') )
plt.plot(df['High'],color = 'blue', label = 'highest values')
plt.xlabel('days')
plt.ylabel('prise')
plt.title('nifty50 chart')
plt.legend(loc=1)
plt.show()
```

```python
# if we have to dates in the space day
v = df['Close']
v.plot(title ='stok market chart',xlabel='date',ylabel='stockprice',
grid=True,figsize=(20,10),fontsize= 18 ,legend=True)
```

BAR PLOT [BAR GRAPH]

```python
df = pd.read_csv("cancer.csv")
df.head()
```

```python
df = df['diagnosis'].value_counts()
df
```

```python
keys = list(df.keys())
values= list(df.values)
plt.bar(keys,values,color=['blue','green'] ,width=0.5,alpha=0.5) #alfa means if we have graphs in transparent form
plt.show()
```

```python
# if we have to plot in horizantel
keys = list(df.keys())
values= list(df.values)
plt.barh(keys,values,color=['blue','green'],alpha=0.5) #alfa means if we have graphs in transparent form
plt.show()

```
